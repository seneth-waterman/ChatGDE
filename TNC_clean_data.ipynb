```python
def HumanRubric(file):
    """Create data frame of human scoring rubric given csv file of scored rubric. 
       Index represents question number

    Arguments:
        file - filepath name of human scored csv file
    Returns:
        human_rubric_df - df of human scored rubric  
    """
    # Create cleaned DF of questions 
    human_rubric_df = pd.read_csv(file)
    human_rubric_df = human_rubric_df.iloc[10:, 3:]
    human_rubric_df = human_rubric_df.reset_index()
    human_rubric_df = human_rubric_df.drop('index', axis=1)
    human_rubric_df.columns = human_rubric_df.iloc[0]
    human_rubric_df = human_rubric_df[1:]

    return human_rubric_df

HumanRubric(file)
```

```python
def ChatEvaluation(file, chat_rubric, a:int=1, b:int=7):
    """ Creates df to represent human answer, chat answer, and correctness of chat's response (0 or 1)

    Arguments:
        file = filepath name of human scored csv file
        chat_rubric = df of chatbot's responses
        a = first question to score
        b = final question to score

    Returns:
        chat_evaluation_df - df of human scored rubric compared to chat scored rubric
    """
    human_rubric = HumanRubric(file)
    chat_evaluation_df = pd.DataFrame({'HumanResponse':human_rubric['Answer'][a-1:b].values,
                                'ChatResponse': chat_rubric['Response'].values}, index=np.arange(a,b+1))
    chat_evaluation_df['AreEqual'] = (chat_evaluation_df['HumanResponse'] == final_score['ChatResponse']).astype(int)
    return chat_evaluation_df

chat_evaluation_df = ChatEvaluation(file, chat_rubric)
chat_evaluation_df
```

```python
def ChatScore(chat_evaluation_df):
    """Percent of chat's scores that are equal to human response

    Arguments: 
        chat_evaluation_df - df of human scored rubric compared to chat scored rubric

    Response:
        score = percent of correct response from chat 
    """
    score = round(100* chat_evaluation_df['AreEqual'].sum() / len(chat_evaluation_df), 2)
    
    return score

ChatScore(chat_evaluation_df)
```
